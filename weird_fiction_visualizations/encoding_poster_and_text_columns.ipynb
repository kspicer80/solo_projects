```python
import pandas as pd
pd.set_option('display.max_rows', 500)
from sklearn.preprocessing import OrdinalEncoder
from pathlib import Path
```

```python
datafile_path = Path(r'.\datasets\fake_dataset_csv.csv')
```

```python
df = pd.read_csv(datafile_path)
```

```python
df.head()
```

```python
df['FullName'] = df[['First Name', 'Last Name']].apply(lambda x: ' '.join(x), axis=1)
df.head()
```

```python
ord_enc = OrdinalEncoder()
df['ordinal_encoded'] = ord_enc.fit_transform(df[['FullName']])
df['ordinal_encoded'] = df['ordinal_encoded'].astype('int')
df.head(10)
```
